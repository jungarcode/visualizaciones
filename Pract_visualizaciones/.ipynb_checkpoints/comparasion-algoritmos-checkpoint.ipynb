```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Ejercicio # 1**

```python
df = pd.read_csv("../datasets/comparasion_algoritmos.csv")
df.head()
```

```python
ax = df.plot.bar(
        x='dataset',
        figsize=(12,5)
)
ax.grid(visible=True)
ax.set_axisbelow(True) # cuadricula detras del grafico
ax.set_title('Comparasion de Algoritmo 1 vs Algoritmo 2')
ax.set_xlabel('Algoritmos')
ax.set_ylabel('Rendimiento de Algoritmos')

ax.legend(labels=['Algoritmo 1','Algoritmo 2'])
```

**Ejecrcicio # 2**

```python
df = pd.read_csv("../datasets/datasets_top_4.csv", index_col= "dataset")
df
```

```python
# Dividimos dataframe

df1= df.iloc[0 : 13, :]
df2= df.iloc[13:, :]
```

```python
print(len(df1))
print(len(df2))
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)

fig.set_size_inches(7, 4)

df1.plot.barh(
    stacked=True,
    ax=ax1,
    width= 0.4,
    color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)
df2.plot.barh(
    stacked=True,
    ax=ax2,
    width= 0.4,
    color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)
ax2.set_ylabel('')
ax1.set_ylabel('Dataset')

ax2.set_xlabel('Porcentaje')
ax1.set_xlabel('Porcentaje')

ax1.legend().remove()

ax2.legend(
    ['Categoria 1','Categoria 2','Categoria 3','Otros'],
    ncol=4,
    loc= "upper left",
    bbox_to_anchor=(-1.2, -0.15)
)

ax2.yaxis.tick_right()

ax1.text(
    75, -5,
    "Algunos datasets han sido ignorados del analisis",
    fontsize=6,
    bbox={
        "facecolor": "orange",
        "alpha": 0.5
    }
)

fig.suptitle('Distribucion de las Categorias\nen los Dataset')
```
