**Ejecrcicio # 2**

Como se distribuyen las diferentes categorias en el dataset

```python
df = pd.read_csv("../datasets/datasets_top_4.csv", index_col= "dataset")
df
```

```python
# Dividimos dataframe

df1= df.iloc[0 : 13, :]
df2= df.iloc[13:, :]
```

```python
print(len(df1))
print(len(df2))
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)

fig.set_size_inches(7, 4)

df1.plot.barh(
    stacked=True,
    ax=ax1,
    width= 0.4,
    color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)
df2.plot.barh(
    stacked=True,
    ax=ax2,
    width= 0.4,
    color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)
ax2.set_ylabel('')
ax1.set_ylabel('Dataset')

ax2.set_xlabel('Porcentaje')
ax1.set_xlabel('Porcentaje')

ax1.legend().remove()

ax2.legend(
    ['Categoria 1','Categoria 2','Categoria 3','Otros'],
    ncol=4,
    loc= "upper left",
    bbox_to_anchor=(-1.2, -0.15)
)

ax2.yaxis.tick_right()

ax1.text(
    75, -5,
    "Algunos datasets han sido ignorados del analisis",
    fontsize=6,
    bbox={
        "facecolor": "orange",
        "alpha": 0.5
    }
)

fig.suptitle('Distribucion de las Categorias\nen los Dataset')
```

**Ejercicio # 3**

Top de paises donde proceden los surfistas

```python
df = pd.read_csv("../datasets/viajes_surfistas.csv")
df
```

```python
df['homecountry'].unique()
```

```python
len(df['homecountry'].unique())
```

```python
cant_surf_pais = df.groupby('homecountry').agg({
    'homename':'count'
})

cant_surf_pais
```

```python
ordenado = cant_surf_pais.sort_values('homename', ascending=False)
ordenado
```

```python
top_4 = ordenado.head(4)
top_4
```

```python
otros = ordenado.drop(index=top_4.index.values)
otros
```

```python
otros['pais'] = 'otros'
otros
```

```python
suma_otros = otros.groupby('pais').agg({
    'homename':'sum'
})

suma_otros
```

```python
suma_otros = suma_otros.reset_index()
suma_otros
```

```python
top_4 = top_4.reset_index()
top_4
```

```python
top_4.columns = ['pais','homename'] # cambio nombre de columna para poder concatenarlo con dataframe otros
top_4
```

```python
df_final = pd.concat([top_4,suma_otros])
df_final
```

```python
df_final = df_final.set_index('pais')
df_final
```

```python
ax = df_final.plot.pie(
    y='homename',
    startangle=90,
    autopct='%1.1f%%',
    colormap='Pastel1'
)

ax.set_ylabel('')
ax.legend(
    title='Pais Surfistas',
    fontsize=10,
    bbox_to_anchor=(1,1)
)

ax.set_title('Porcentaje de Surfistas por Pais')
```

**Ejemplo # 4**

Tiendas con mayor numero de ventas y como ha sido su evolucion en el tiempo

```python
df = pd.read_csv('../datasets/tienda_ventas.csv')
df
```

```python
len(df['store_nbr'].unique())
```

```python
len(df['date'].unique())
```

```python
df['mes'] = df['date'].apply(lambda x: x[:7])
df
```

```python
len(df['mes'].unique())
```

```python
agrupado_tda = df.groupby('store_nbr').agg({
    'sales':'sum'
})

agrupado_tda
```

```python
agrupado_tda.describe()
```

```python
agrupado_tda.plot.box()
```

```python
agrupado_tda.plot.hist(bins=20)
```

```python
agrupado_tda.sort_values('sales', ascending=False)
```

```python
top_5_tdas = [44,45,47,3,46]
```

```python
df_top5 = df[df['store_nbr'].isin(top_5_tdas)]
df_top5
```

```python
top_5_mes = df_top5.groupby(['store_nbr', 'mes']).agg({
    'sales':'sum'
})
top_5_mes = top_5_mes.reset_index()
top_5_mes
```

```python
ax = sns.lineplot(
    data = top_5_mes,
    x = 'mes',
    y = 'sales',
    hue = 'store_nbr',
    style = 'store_nbr',
    markers = True,
    palette = 'Set1'
)
ax.set_title('Ventas por mes TOP 5 Tiendas')
ax.tick_params(
    axis = 'x',
    labelrotation = 90,
    labelsize = 8
)
ax.tick_params(
    axis = 'y',
    labelsize = 8
)

ax.grid(
    visible = True,
    axis = 'x',
    alpha = 0.5
)
ax.grid(
    visible = True,
    axis = 'y',
    alpha = 0.5
)

ax.legend(
    title = 'Top 5 Tiendas',
    fontsize = 8
)

ax.set_ylabel('Ventas')
ax.set_xlabel('Mes')

fig = ax.get_figure()
fig.set_size_inches(11,3)
```
